```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importing the dataset

#dfAccData4 = pd.read_csv('exp2dataXYZ.csv')
#dfAccData4 = pd.read_csv('exp3dataXYZ.csv')
dfAccData4 = pd.read_csv('exp4dataXYZ.csv')
```

```python
dfAccData4
```

```python
# Generating gradient columns

X_list = [x for x in dfAccData4["X"]]
X_list_copy = [y for y in dfAccData4["X"]]
X_list_copy.pop(0)
X_list_copy.insert(-1, X_list[-1])

X_grad = [(y-x) for y,x in zip(X_list_copy, X_list)]
X_grad.insert(0, 0.00)
X_grad.pop(-1)

X_gradf = [x if abs(x) > 0.02 else 0.00 for x in X_grad]
dfAccData4["X_grad"] = X_gradf


Y_list = [x for x in dfAccData4["Y"]]
Y_list_copy = [y for y in dfAccData4["Y"]]
Y_list_copy.pop(0)
Y_list_copy.insert(-1, Y_list[-1])

Y_grad = [(y-x) for y,x in zip(Y_list_copy, Y_list)]
Y_grad.insert(0, 0.00)
Y_grad.pop(-1)

Y_gradf = [y if abs(y) > 0.02 else 0.00 for y in Y_grad]
dfAccData4["Y_grad"] = Y_gradf


Z_list = [x for x in dfAccData4["Z"]]
Z_list_copy = [y for y in dfAccData4["Z"]]
Z_list_copy.pop(0)
Z_list_copy.insert(-1, Z_list[-1])

Z_grad = [(y-x) for y,x in zip(Z_list_copy, Z_list)]
Z_grad.insert(0, 0.00)
Z_grad.pop(-1)

Z_gradf = [z if abs(z) > 0.02 else 0.00 for z in Z_grad]
dfAccData4["Z_grad"] = Z_gradf


pd.set_option('display.max_rows', None)
dfAccData4
```

```python
# Generating laplacian columns

X2_list = [x for x in dfAccData4["X_grad"]]
X2_list_copy = [y for y in dfAccData4["X_grad"]]
X2_list_copy.pop(0)
X2_list_copy.insert(-1, X2_list[-1])

X2_grad = [(y-x) for y,x in zip(X2_list_copy, X2_list)]
X2_grad.insert(0, 0.00)
X2_grad.pop(-1)

dfAccData4["X_laplacian"] = X2_grad

Y2_list = [x for x in dfAccData4["Y_grad"]]
Y2_list_copy = [y for y in dfAccData4["Y_grad"]]
Y2_list_copy.pop(0)
Y2_list_copy.insert(-1, Y2_list[-1])

Y2_grad = [(y-x) for y,x in zip(Y2_list_copy, Y2_list)]
Y2_grad.insert(0, 0.00)
Y2_grad.pop(-1)

dfAccData4["Y_laplacian"] = Y2_grad

Z2_list = [x for x in dfAccData4["Z_grad"]]
Z2_list_copy = [y for y in dfAccData4["Z_grad"]]
Z2_list_copy.pop(0)
Z2_list_copy.insert(-1, Z2_list[-1])

Z2_grad = [(y-x) for y,x in zip(Z2_list_copy, Z2_list)]
Z2_grad.insert(0, 0.00)
Z2_grad.pop(-1)

dfAccData4["Z_laplacian"] = Z2_grad

dfAccData4
```

```python
# Plotting

x1 = np.array(dfAccData4["Time"])
y1 = [np.array(dfAccData4["X_grad"]), np.array(dfAccData4["Y_grad"]), np.array(dfAccData4["Z_grad"])]
y1_ = np.transpose(y1)

x2 = np.array(dfAccData4["Time"])
y2 = [np.array(dfAccData4["X_laplacian"]), np.array(dfAccData4["Y_laplacian"]), np.array(dfAccData4["Z_laplacian"])]
y2_ = np.transpose(y2)

fig, (ax1, ax2) = plt.subplots(2, figsize = (20,10))
fig.suptitle('Gradient and Laplacian')
ax1.plot(x1, y1_)
ax2.plot(x2, y2_)
```

```python
# Fourier Transform
#https://towardsdatascience.com/fourier-transform-for-time-series-292eb887b101

Y_ = np.array(dfAccData4["Y_grad"])
Yf = abs(np.fft.fft(Y_))

X_ = list(range(len(dfAccData4.index)))
num = np.size(X_)
freq = [i / num for i in list(range(num))]

spectrum = Yf.real*Yf.real + Yf.imag*Yf.imag
nspectrum = spectrum/spectrum[0]   # normalized_spectrum

plt.plot(freq,spectrum)  # (freq,nspectrum)
```

```python
# Another Fourier Transform
#https://fischerbach.medium.com/introduction-to-fourier-analysis-of-time-series-42151703524a
#https://colab.research.google.com/github/FabrizioMusacchio/Python_Neuro_Practical/
# blob/master/05%20Using%20Fourier%20transform%20for%20time%20series%20deconstruction%20.ipynb

import scipy.fft
A_signal_fft = scipy.fft.rfft(Y_)
frequencies = scipy.fft.rfftfreq(num, 1/263)

fig=plt.figure(2, figsize=(15,6))
plt.clf()
plt.plot(frequencies, np.abs(A_signal_fft), lw=1.0, c='paleturquoise')
plt.stem(frequencies, np.abs(A_signal_fft))
```

```python
# Reducing 'spectral leakage'
#http://qingkaikong.blogspot.com/2016/10/signal-processing-why-do-we-need-taper.html
#https://www.zhinst.com/others/en/blogs/how-control-spectral-leakage-window-functions-labone

# We will use (blackman) window(as a taper) to isolate the dominant frequency
#https://docs.scipy.org/doc/scipy/reference/signal.windows.html

from scipy import signal

window = signal.windows.blackman(263)
plt.plot(window)
plt.title("Window")
plt.ylabel("Amplitude")
plt.xlabel("Sample")
```

```python
# Applying this window

Y_taper = Y_ * window

plt.plot(x1,Y_taper)
```

```python
B_signal_fft = scipy.fft.rfft(Y_taper)
frequencies = scipy.fft.rfftfreq(num, 1/263)

fig=plt.figure(2, figsize=(15,6))
plt.clf()
plt.plot(frequencies, np.abs(B_signal_fft), lw=1.0, c='paleturquoise')
plt.stem(frequencies, np.abs(B_signal_fft))
```
