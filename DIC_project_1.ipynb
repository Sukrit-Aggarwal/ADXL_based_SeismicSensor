```python
# Fourier Transform

Y_ = np.array(dfAccData4["Y"])
Yf = abs(np.fft.fft(Y_))

X_ = list(range(len(dfAccData4.index)))
num = np.size(X_)
freq = [i / num for i in list(range(num))]

spectrum = Yf.real*Yf.real + Yf.imag*Yf.imag
nspectrum = spectrum/spectrum[0]   # normalized_spectrum

plt.plot(freq,spectrum)  # (freq,nspectrum)
```
